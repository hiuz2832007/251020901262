import math

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a == 0:
    if b == 0:
        print("Vô số nghiệm." if c == 0 else "Vô nghiệm.")
    else:
        print("Nghiệm của phương trình bậc nhất:", -c/b)
else:
    delta = b*b - 4*a*c
    print("Delta =", delta)

    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Hai nghiệm phân biệt:")
        print("x1 =", x1)
        print("x2 =", x2)
    elif delta == 0:
        print("Nghiệm kép x =", -b/(2*a))
    else:
        print("Phương trình vô nghiệm thực.")
